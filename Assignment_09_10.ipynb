```python
import numpy as np
import matplotlib.pyplot as ply
import seaborn as sns
d=sns.load_dataset('diamonds')
d
```

```python
sns.histplot(data=d,y='carat',x='price',hue='cut')
```

```python
sns.jointplot(data=d,y='carat',x='price',hue='color')
```

```python
sns.kdeplot(data=d,x='price',y='carat',hue='clarity')
```

```python
sns.pairplot(data=d[['carat','price','clarity']],hue='clarity')
```
